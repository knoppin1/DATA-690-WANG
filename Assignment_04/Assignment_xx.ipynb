---
Read (with naked eyes, not computer) the dictionary and find out the definitions of the following 12 columns and copy the text next to the column name.
*   ***UNITID***:  Unique identification number assigned to postsecondary institutions
*   ***INSTNM***:  Name of institution
*   ***CITY***: City of institution's location
*   ***ZIP***: Zip code of institution's location
*   ***STABBR***: State abreviation of institution's location
*   ***ADM_RATE***: Fall Admission rate for institution. Number of admitted undergraduates divided by number of applicants
*   ***UGDS***: Number of degree/certificate-seeking undergraduates enrolled in the fall
*   ***TUITIONFEE_IN***: In-state tuition fees
*   ***MN_EARN_WNE_P10***: Mean earnings for institutional aggregate of all federally aided students who enroll in an institution each year and who are
employed but not enrolled.  This mean reflects 10 years after enrollment.



```
# Re-import the data with above 9 columns (ignore all other variables)
columns_of_interest = ['UNITID', 'INSTNM', 'CITY', 'ZIP', 'STABBR', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10']
colleges_df = pd.read_csv('MERGED2017_18_PP.csv', header = 0, usecols = columns_of_interest)  
colleges_df.shape
```

```
colleges_df.head(4).T
```

```
# Find out the 10 colleges that are hardest to get in (top 10)
colleges_df.sort_values(by=['ADM_RATE','UGDS'], ascending=True).head(10)
```

```
# Find out the average in-state tuition of top 10
top10_expensive_df = colleges_df.sort_values(by=['TUITIONFEE_IN'], ascending=False).head(10)
top10_expensive_df["TUITIONFEE_IN"].mean()
```

```
# Find out the average earnings of graduates from the top 10
top10_earnings_df = colleges_df.sort_values(by=['MN_EARN_WNE_P10'], ascending=False).head(10)
top10_earnings_df["MN_EARN_WNE_P10"].mean()
```

```
# Find out the 10 colleges that are easiest to get in (bottom 10)
colleges_df.sort_values(by=['ADM_RATE','UGDS'], ascending=False).head(10)
```

```
# Find out the average in-state tuition of the bottom 10
bot10_expensive_df = colleges_df.sort_values(by=['TUITIONFEE_IN'], ascending=True).head(10)
bot10_expensive_df["TUITIONFEE_IN"].mean()
```

```
# Find out the average earnings of graduates from the bottom 10
bot10_earnings_df = colleges_df.sort_values(by=['MN_EARN_WNE_P10'], ascending=True).head(10)
bot10_earnings_df["MN_EARN_WNE_P10"].mean()
```
