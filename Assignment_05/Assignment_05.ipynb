## Documented definitions for the following 9 columns of the data set.
*   ***UNITID***:  Unique identification number assigned to postsecondary institutions
*   ***INSTNM***:  Name of institution
*   ***CITY***: City of institution's location
*   ***ZIP***: Zip code of institution's location
*   ***STABBR***: State abreviation of institution's location
*   ***ADM_RATE***: Fall Admission rate for institution. Number of admitted undergraduates divided by number of applicants
*   ***UGDS***: Number of degree/certificate-seeking undergraduates enrolled in the fall
*   ***TUITIONFEE_IN***: In-state tuition fees
*   ***MN_EARN_WNE_P10***: Mean earnings for institutional aggregate of all federally aided students who enroll in an institution each year and who are
employed but not enrolled.  This mean reflects 10 years after enrollment.

### Find and display UMBC

```
umbc = colleges_df[colleges_df['INSTNM'] == 'UMBC']
umbc
```

Note: No results with 'UMBC', so expand the acronym to substring search for the words in the institution name

```
umbc = colleges_df[colleges_df['INSTNM'].str.contains('Maryland') & colleges_df['INSTNM'].str.contains('Baltimore County')]
umbc
```

### Find and display Univ. of Maryland College Park

```
umcp = colleges_df[colleges_df['INSTNM'].str.contains('Maryland') & colleges_df['INSTNM'].str.contains('College Park')]
umcp
```

### Find and display Loyola College Maryland

```
loyola = colleges_df[colleges_df['INSTNM'].str.contains('Maryland') & colleges_df['INSTNM'].str.contains('Loyola')]
loyola
```

 ### Find and display Johns Hopkins

```
jhu = colleges_df[colleges_df['INSTNM'].str.contains('Johns Hopkins') ]
jhu
```

### Calculate the min, max, and average in-state tution of the above four institutions.

```
# Append dataframes 
df = umbc.append(umcp,ignore_index=True).append(loyola,ignore_index=True).append(jhu,ignore_index=True)
df 
```

Minimum tuition

```
min_tuition = min(df['TUITIONFEE_IN'])
min_tuition
```

```
# Display intstitution with minimum tuition
# Reference: https://pythonexamples.org/pandas-dataframe-query/
filter = "TUITIONFEE_IN == " + str(min_tuition)
df1 = df.query(filter)
print(f"{df1.iloc[0,1]} has minimum tuition of {'${:,.2f}'.format(min_tuition)}")
```

Maximum tuition

```
max_tuition = max(df['TUITIONFEE_IN'])
max_tuition
```

```
# Display intstitution with maximum tuition
institution = df[df.TUITIONFEE_IN.eq(max_tuition)].iloc[0,1]
print(f"{institution} has maximum tuition of {'${:,.2f}'.format(max_tuition)}")
```

Average in-state tuition

```
avg_tuition = df["TUITIONFEE_IN"].mean()
avg_tuition
```

```
print(f"Average in-state tuition for selected colleges is {'${:,.2f}'.format(avg_tuition)}")
```
